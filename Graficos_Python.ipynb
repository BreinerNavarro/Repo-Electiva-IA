```python
# Importacion de librerias
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
# Grafica de linea con Matplotlib
x = np.linspace(-5, 5, 100)
y = x**2

plt.plot(x,y, color = "purple", label = "y = x^2", linewidth = 2)
plt.title("Función Cuadratica")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

```python
x = [2, 4, 6, 7, 10, 15]       # numero de horas de estudio
y = [10, 23, 45, 68, 81, 92]   # punteje de cada estudiante de cero a cien
 
plt.scatter(x,y, s = 100, color = "r", marker = '^' , alpha=0.8 )
plt.title("Horas de estudio VS Puntaje en un examen")
plt.xlabel("Horas de estudio (horas)")
plt.ylabel("Puntaje (0-100)")
plt.grid()
```

```python
#Grafica de dispersión con Plotly
df = px.data.iris() # Dataset Iris de ejemplo
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 title="Dispersión: Longitud vs Ancho de Sépalo")
fig.show()
```

```python
# Grafico de lineas bacano
plt.style.use('dark_background') # Tema oscuro

# Generar el dataset de ondas
x = np.linspace(0, 2, 800) 
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

# Definimos colores específicos
colors = ['#1f77b4', '#ff7f0e', "#030303", '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']
markers = ['>', '>','>','>','>','>','>','>']

# Dibujar las curvas con efectos
for i in range(8):
    # Parámetros variables para cada onda
    amp = 8 - i  # Amplitudes decrecientes para que no se amontonen
    freq = 1.5 + (i * 0.2)
    phase = i * 0.8
    
    y = amp * np.sin(2 * np.pi * freq * x + phase)
    
    # Dibujamos una sombra ligera debajo de la línea para dar profundidad
    ax.plot(x, y, color=colors[i], linewidth=2, label=f'y{i}(x)',
            marker=markers[i], markevery=40, markersize=6, 
            alpha=0.9, path_effects=[]) # path_effects se puede usar para sombras

# Refinar la cuadrícula y etiquetas
ax.set_title("tracefig - Multi plots", fontsize=15, pad=15, color='white')
ax.set_xlabel("x [s]", fontsize=12)
ax.set_ylabel("y [SI]", fontsize=12)

# Cuadrícula personalizada (Grid)
ax.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.5)

# Ajustar los límites para que se vea limpio
ax.set_xlim(0, 2)
ax.set_ylim(-9, 9)

# Leyenda
ax.legend(loc='upper left', frameon=True, shadow=True, facecolor='#333333', edgecolor='none')

# Mostrar resultado
plt.tight_layout()
plt.show()
```
