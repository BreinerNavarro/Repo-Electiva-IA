```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = x**2

plt.plot(x,y, color = "purple", label = "y = x^2", linewidth = 2)
plt.title("Función Cuadratica")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.legend()
plt.show()
```

```python
x = [2, 4, 6, 7, 10, 15]       # numero de horas de estudio
y = [10, 23, 45, 68, 81, 92]   # punteje de cada estudiante de cero a cien
 
plt.scatter(x,y, s = 100, color = "r", marker = '^' , alpha=0.8 )
plt.title("Horas de estudio VS Puntaje en un examen")
plt.xlabel("Horas de estudio (horas)")
plt.ylabel("Puntaje (0-100)")
plt.grid()
```
